```python
import numpy
import torch
import matplotlib.pyplot as plt
from tropix_linear import TropixLinear
```

```python
tpx = TropixLinear(2, 3)

[X1, X2] = numpy.meshgrid(numpy.linspace(-10, 10, 100), numpy.linspace(-10, 10, 100))
input = torch.tensor(numpy.column_stack((
    X1.flatten(), X2.flatten()
)), dtype=torch.float32)

tpx.set_parameters(torch.tensor(numpy.array(
    [[0.1, 0.2], [-0.1, -0.2], [0.1, -0.2]]
)), bias=torch.tensor([0.0, 0.0, 0.0]))

r = tpx.forward(input)

plt.figure()
plt.grid()
plt.contour(X1, X2, numpy.reshape(r.detach()[:, 0].numpy(), X1.shape), levels=20)
plt.show()

tpx.behave_as_tropix = False
r = tpx.forward(input)

plt.figure()
plt.grid()
plt.contour(X1, X2, numpy.reshape(r.detach()[:, 0].numpy(), X1.shape), levels=20)
plt.show()
```

```python
tpx = TropixLinear(1, 1)
X1 = numpy.linspace(-10, 10, 100)
input = torch.tensor(numpy.column_stack((X1.flatten(), )), dtype=torch.float32)

tpx.set_parameters(
    torch.tensor([[0.5]]), bias=torch.tensor([-1])
)

r = torch.relu(tpx.forward(input))

plt.figure()
plt.grid(True)
plt.plot(X1, r.detach().numpy())
plt.show()
```
